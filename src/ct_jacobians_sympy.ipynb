```python
import sympy as sp

zeta_g, eta_g, theta_g, phi_g = sp.symbols('zeta_g, eta_g, theta_g, phi_g')
zeta_a, eta_a, theta_a, omega_a = sp.symbols('zeta_a, eta_a, theta_a, omega_a')
vg, va, L = sp.symbols('v_g, v_a, L')

x = sp.Matrix([[zeta_g], [eta_g], [theta_g]])
u = sp.Matrix([[vg], [phi_g]])

eqn1 = vg*sp.cos(theta_g)
eqn2 = vg*sp.sin(theta_g)
eqn3 = (vg/L)*sp.tan(phi_g)

F = sp.Matrix([[eqn1], [eqn2], [eqn3]])
```

```python
# compute the A-bar
F.jacobian(x)
```

```python
# compute the B-bar
F.jacobian(u)
```

```python


xa = sp.Matrix([[zeta_a], [eta_a], [theta_a]])
ua = sp.Matrix([[va], [omega_a]])

eqn1 = va*sp.cos(theta_a)
eqn2 = va*sp.sin(theta_a)
eqn3 = omega_a

F = sp.Matrix([[eqn1], [eqn2], [eqn3]])
```

```python
F.jacobian(xa)
```

```python
F.jacobian(ua)
```
